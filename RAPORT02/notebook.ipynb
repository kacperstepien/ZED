## Import bibliotek

```python
import pandas as pd
from sklearn import linear_model
#wykomentowane bibilioteki których nie użyłem ostatecznie
#from mlxtend.regressor import StackingRegressor
#from sklearn import svm
#from sklearn import ensemble
```

## Odczyt danych treningowych

```python
data_frame = pd.read_csv("train.csv", sep=",", dtype='unicode')
#próbowałem zamiany daty na wartość liczbową i dołożenie jej do modelu, lecz to pogarsza wynik
data_frame.drop('data',1, inplace=True)
# próbowałem wybrania tylko tych kolumn które wykazywały korelację w pierwszym raporcie z R, ale to pogorszyło wynik
#train_df_x = data_frame.loc[: ,["idsito","lat","lon","ora","azimuthi","humidity","temperatura_ambiente","tempi","cloudcover","irri_pvgis_mod","irr_pvgis_mod","irradiamento"]]
train_df_x = data_frame.loc[: ,"id":"irri_pvgis_mod"]
train_df_y = data_frame.loc[: ,"kwh"]
train_df_x
```

## Odczytanie danych testowych

```python
test_df_x = pd.read_csv("test.csv", sep=",", dtype='unicode')
test_df_x.drop('data',1, inplace=True)
pred_df_id = test_df_x.loc[: ,"id"].tolist()
# próbowałem wybrania tylko tych kolumn które wykazywały korelację w pierwszym raporcie z R, ale to pogorszyło wynik
#test_df_x = test_df_x.loc[: ,["idsito","lat","lon","ora","azimuthi","humidity","temperatura_ambiente","tempi","cloudcover","irri_pvgis_mod","irr_pvgis_mod","irradiamento"]]

test_df_x
```

## Utworzenie regresora

```python
#próbowałem użyć także innych modeli, ale okazały się gorsze

#logr = linear_model.LogisticRegression()
#br = linear_model.BayesianRidge()
#svr = svm.SVR()
#rf = ensemble.RandomForestRegressor()
#regr = StackingRegressor(regressors=[svr, br, rf], 
#                           meta_regressor=logr)
regr = linear_model.LinearRegression()
regr.fit(train_df_x, train_df_y)
regr
```

## Predykcja i zapis wyniku

```python
pred_df_y = regr.predict(test_df_x)

# tutaj co ciekawe, dla wartości poniżej zera lepsze wyniki dostaje sie przy ich wartości bezwzględnej, a nie przy zamianie na 0
pred_df_y = [abs(x) for x in pred_df_y]

d = {'Id' : pred_df_id,
    'kwh' : pred_df_y}
pred_df = pd.DataFrame(d)

pred_df.to_csv('prediction.csv',index=False)

pred_df
```
